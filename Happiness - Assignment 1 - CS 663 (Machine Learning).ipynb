# Question 3
* Of the countries in this dataset, which country had the largest increase in happiness from its start
of participation to 2023?
    * answer: Congo (Brazzaville)
* Which had the largest decrease in happiness?
    * answer: Lebanon
    

```python
first_year_happiness = df.groupby('Country name')['Life Ladder'].first().reset_index(name='First Year Happiness')
last_year_happiness = df[df['year'] == 2022].groupby('Country name')['Life Ladder'].last().reset_index(name='Last Year Happiness')
happiness_change = pd.merge(first_year_happiness, last_year_happiness, on='Country name')
happiness_change['Happiness Change'] = happiness_change['Last Year Happiness'] - happiness_change['First Year Happiness']
largest_increase_country = happiness_change[happiness_change['Happiness Change'] == happiness_change['Happiness Change'].max()]
print(largest_increase_country[['Country name', 'Happiness Change']])
```
